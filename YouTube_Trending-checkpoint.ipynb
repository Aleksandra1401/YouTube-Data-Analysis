# **Preprocessing**

Define a function to take the country name and load appropriate .json file for the country. Get the category id and title from json and create the dataframe containing these fields. Return this dataframe.

```python
def get_category_df(country):
  filename = country + '_category_id.json'
  categories = open(filename)
  categories_json = json.load(categories)
  categories_dict = {int(item['id']):item['snippet']['title'] for item in categories_json['items']}
  categories = pd.DataFrame(list(categories_dict.items()), columns = ['category_id','category_title'])
  return categories
```

Load all datasets and create category dataframe for each country dataframe.

```python
CA_data = pd.read_csv('CAvideos.csv')
DE_data = pd.read_csv('DEvideos.csv')
FR_data = pd.read_csv('FRvideos.csv')
GB_data = pd.read_csv('GBvideos.csv')
IN_data = pd.read_csv('INvideos.csv')
# 4 datasets below have encoding issues that I did not solve, so these will be excluded from analisys.
#MX_data = pd.read_csv('MXvideos.csv')
#JP_data = pd.read_csv('JPvideos.csv')
#KR_data = pd.read_csv('KRvideos.csv')
#RU_data = pd.read_csv('RUvideos.csv')
US_data = pd.read_csv('USvideos.csv')
CA_categories = get_category_df('CA')
DE_categories = get_category_df('DE')
FR_categories = get_category_df('FR')
GB_categories = get_category_df('GB')
IN_categories = get_category_df('IN')
US_categories = get_category_df('US')
```

Merge datasets for country data with appropriate category dataframe to add ordinal category names to each dataframe.

```python
CA_data = pd.merge(CA_data,CA_categories, how = 'left', on = 'category_id')
DE_data = pd.merge(DE_data,DE_categories, how = 'left', on = 'category_id')
FR_data = pd.merge(FR_data,FR_categories, how = 'left', on = 'category_id')
GB_data = pd.merge(GB_data,GB_categories, how = 'left', on = 'category_id')
IN_data = pd.merge(IN_data,IN_categories, how = 'left', on = 'category_id')
US_data = pd.merge(US_data,US_categories, how = 'left', on = 'category_id')
```

Fill nans in the 'category_title' with the value with the most frequent occurence for each dataset. Define the function to return the mode value and apply on each dataset.

```python
def get_mode(df, column_name):
  return df[column_name].mode().values[0]
```

```python
# get the most frequent values
CA_mode = get_mode(CA_data, 'category_title')
FR_mode = get_mode(FR_data, 'category_title')
DE_mode = get_mode(DE_data, 'category_title')
GB_mode = get_mode(GB_data, 'category_title')
IN_mode = get_mode(IN_data, 'category_title')

#fill the dataframes with mode values for category_title
CA_data['category_title'].fillna(value = CA_mode,inplace = True)
FR_data['category_title'].fillna(value = FR_mode,inplace = True)
DE_data['category_title'].fillna(value = DE_mode,inplace = True)
GB_data['category_title'].fillna(value = GB_mode,inplace = True)
IN_data['category_title'].fillna(value = IN_mode,inplace = True)

#check
print(CA_data.columns[CA_data.isnull().any()]) #let's check results
print(FR_data.columns[FR_data.isnull().any()])
print(DE_data.columns[DE_data.isnull().any()])
print(GB_data.columns[GB_data.isnull().any()])
print(IN_data.columns[IN_data.isnull().any()])
```

Concat country dataframes and drop columns that are not needed. Cast columns representing dates to datetime data type, and other ordinal columns representing titles to strings. 

```python
countries_df = pd.concat([CA_data,DE_data,GB_data,FR_data,IN_data,US_data])
countries_df = countries_df.drop(['category_id','thumbnail_link','description'],axis=1)
countries_df['video_id'] = countries_df['video_id'].astype('string')
countries_df['title'] = countries_df['title'].astype('string')
countries_df['category_title'] = countries_df['category_title'].astype('string')
countries_df['trending_date'] = pd.to_datetime(countries_df['trending_date'], format='%y.%d.%m')
countries_df["publish_time"]=pd.to_datetime(countries_df["publish_time"])
countries_df = countries_df.reset_index(drop=True)
print(countries_df.shape)
countries_df.dtypes
```

To find out how many views, likes, dislikes and comments each category has, we can make a function that will take the list of categories and traget column for which we need answers. Function will return series holding the sums of the values in column with respect to all categories. Sums are sorted in descending order.

```python
def get_sum_by_category(categories_list, target_column):
  grouped = countries_df.groupby('category_title')[target_column]
  result_values = []
  for category in categories_list:
   result_values.append(grouped.get_group(category).sum())
  series = pd.Series(result_values, index = categories)
  series = series.sort_values(ascending = False)
  return series
```

```python
grouped = countries_df.groupby('video_id')
video_names = []
video_views = []
for name, group in grouped:
    video_names.append(grouped.get_group(name).iloc[0]['title'])
    video_views.append(grouped.get_group(name)['views'].max())

```

Show 15 most viewed videos. Interestingly, it seems that pop music is the most viewed content, and Latin American authors are more popular than I thought.

```python
zipped = list(zip(video_names,video_views))
average_video_views = pd.DataFrame(zipped, columns=['title', 'views'])
average_video_views.rename(columns = {0:'views'}, inplace = True)
average_video_views.sort_values(by='views',ascending = False,inplace=True)
average_video_views.reset_index(drop=True,inplace=True)
average_video_views.head(15)
```

Now, I was really curious what are some most liked and disliked(especially) videos in the dataset. But, we would make a mistake if we answered this question solely based on raw numbers in 'likes' and 'dislikes' columns. Every video has different amount of views and therefore, we should think of this problem having that in mind.  
  
My idea is to consider the difference in likes and dislikes a variable that can tell me how many more likes than dislikes video has and the other way around. I will construct a column for the values of likes-dislikes and dislikes-likes to find out the answer. Values will be sorted in descending order so we can get top results.

```python
differences_df = countries_df[['title','views','likes','dislikes']].copy(deep = True)
differences_df.drop_duplicates(inplace=True)
differences_df = pd.merge(average_video_views,differences_df, how = 'left', on=['title','views'])
print(differences_df.shape)
differences_df.isnull().sum()
```

I am not sure how nans appeared, but comparing to the amount of data we have all the information for, the amount of nans isn't as large. I'll drop them.

```python
differences_df.dropna(inplace = True)
differences_df.shape
```

Now we can add columns for likes vs dislikes and dislikes vs likes

```python
differences_df['liked'] = differences_df['likes'] - differences_df['dislikes']
differences_df['disliked'] = differences_df['dislikes'] - differences_df['likes']
differences_df.sort_values(by = 'liked',ascending=False)[['title','liked']].head(15)
```

The description on kaggle.com said that there were comments for videos included in the dataset, but this is not true. Therefore, I will explore what I have for text data, which is 'tags' column. We can get the words that are most frequent among trending tags.
I used NLP technique to get frequency distribution of most common words for tags by example from  
https://realpython.com/nltk-nlp-python/#making-a-frequency-distribution

```python
countries_df['tags'] = countries_df['tags'].astype('string')
countries_df['tags'].head()
```

```python
tags_string = ' '.join(countries_df["tags"])
print(tags_string[0: 500])
```

```python
tags_string = tags_string.replace("|", " ")
tags_string = tags_string.replace("\"", "")
print(tags_string[0:500])
```

```python
tags_string = ' '.join( [w for w in tags_string.split() if len(w)>1] ) #let's get rid of individual characters if any
print(tags_string[0:300])
```

```python
from nltk import FreqDist
import nltk
nltk.download('punkt')


words = nltk.tokenize.word_tokenize(tags_string)
frequency_distribution = FreqDist(words)
frequency_distribution.plot(40, cumulative=True)
```

```python
frequency_distribution.most_common(100)
```
